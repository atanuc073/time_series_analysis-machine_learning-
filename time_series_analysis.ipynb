## Plot Autocorelation Function (ACF)



```python
#Acf of s&p 
sgt.plot_acf(df.market_values,lags=40,zero=False)
plt.title('ACF of S&P', size=24)
```

```python
#ACf of white noise
#remember white noise is never similar to previous
sgt.plot_acf(df.wn,zero=False,lags=40)
plt.title('ACF of White Noise',size=24)
```

```python
sgt.plot_acf(df.rw,zero=False,lags=40)
plt.title('ACF of Random Walk',size=24)
```

If we want calculate direct relation of a time series with its lag version the we need to measure Partial Autocorrelation Function (PACF)

```python
sgt.plot_pacf(df.market_values,lags=40,zero=False,method='ols')
plt.title('PACF of s&p',size=24)
```

```python
#Pacf of White Noise
sgt.plot_pacf(df.wn,lags=40,zero=False,method='ols')
plt.title('PACF of wn',size=24)
```
